# Olympics Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summer.csv")
df
```

## 1. In how many cities Summer Olympics is held so far?

```python
data = df['Country'].unique()
len(data)
```

## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = df[df['Medal'] == 'Gold'].groupby(['Sport']).first().reset_index()
data['Sport'].head()
```

## 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for sport in df['Sport'].unique():
    data.append([sport , len(df[df['Sport']  == sport])])

pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head()
```

## 4. Which player has won most number of medals? (Top 5)

```python
data = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete , len(df[df['Athlete']  == Athlete])])

pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head()
```

## 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete , len(df[df['Athlete']  == Athlete])])

d = pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False)
d.head()
```

## 6. In which year India won first Gold Medal in Summer Olympics?

```python
data = []

for medal in df['Medal'].unique():
    data.append([medal , len(df[df['Medal']  == medal])])

pd.DataFrame(data,columns = ['Medal','Country']).sort_values(by='Medal', ascending=False)
```

```python
data = df[df['Medal'] == 'Gold'].groupby(['Year']).first().reset_index()
data['Country']
```

## 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Athlete in df['Event'].unique():
    data.append([Athlete , len(df[df['Athlete']  == Athlete])])

pd.DataFrame(data,columns = ['Athlete','Event']).sort_values(by='Event', ascending=False)
```

```python
data = []

for medal in df['Medal'].unique():
    data.append([medal , len(df[df['Medal']  == medal])])

pd.DataFrame(data,columns = ['Medal','Sport']).sort_values(by='Medal', ascending=False)
```
